### Q1-5. 월별 카테고리별 태그 키워드 순위


태그 데이터가 Nan이 아닌 값만 추출한다.

````python
have_tags = df.loc[df['tags'].notnull()].copy()
len(have_tags)
````

태그의 공백을 제거하고 모든 영문자를 소문자로 치환한다.

````python
have_tags['tags'] = have_tags['tags'].str.replace(' ', '')
have_tags['tags'] = have_tags['tags'].str.lower()
````

````python
tag_month_rank = have_tags.groupby([have_tags['published_date'].dt.month, 'category_name'])['tags'].sum()
tag_month_rank = tag_month_rank.unstack()

tag_list = tag_month_rank.columns.tolist()
strings = []
for month,rows in tag_month_rank.iterrows():
    strings.append(rows.str.split('|'))

results = []
for idx1 in range(len(strings)):
    for idx2 in range(len(tag_list)):
        if type(strings[idx1][idx2]) == float:
            results.append(0)
        else:
            c = Counter(strings[idx1][idx2]).most_common(10)
            results.append(c)

month_idx = 3
for idx, rows in enumerate(results):
    if idx % 15 == 0:
        print(f'\n{month_idx}월')
        month_idx += 1

    if rows == 0:
        print(f'[{tag_list[idx%15]}] : 태그 없음')
    else:
        res = []
        for row in rows:
            res.append(f'{row[0]}({row[1]})')
        print(f'[{tag_list[idx%15]}] : {res}')
````

월별 가장 많이 등장한 순서대로 태그를 출력했고 괄호 안의 숫자는 월별 태그 등장 횟수를 의미한다.

더 정확한 태그 키워드 순위를 알고자 한다면 문자열의 더 많은 전처리가 필요하다. 예를 들자면 브이로그와 vlog는 의미가 같으므로 브이로그를 vlog로 치환하는 전처리가 필요하다.

또 최근 유튜브 제목만으로 동영상 내용을 파악하기 어렵기 때문에 인기동영상 종류를 파악하는데 카테고리와 태그를 활용하는 것이 좋아보인다.
